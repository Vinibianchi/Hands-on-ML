```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
np.random.seed(4)
m = 60
w1, w2 = 0.1, 0.3
noise = 0.1

angles = np.random.rand(m) * 3 * np.pi / 2 - 0.5
X = np.empty((m, 3))
X[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * np.random.randn(m) / 2
X[:, 1] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 2] = X[:, 0] * w1 + X[:, 1] * w2 + noise * np.random.randn(m)
```

```python
X.shape
```

```python
x =X[:,0]
y =X[:,1]
z =X[:,2]

#Criando a Figura
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection = "3d")

# plotando os dados

ax.scatter3D(x,y,z, color = 'green')
plt.title('Dataset em 3D')
plt.show()
```

# PCA

```python
X_centered = X - X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_centered)
```

```python
Vt.T
```

```python
c1 = Vt.T[:0] # Vetor PC1
c2 = Vt.T[:1] # Vetor PC2
```

```python
m, n = X.shape

S= np.zeros(X_centered.shape)
S[:n, :n] = np.diag(s)
```

## Projetando o conjunto de treino em 2D

```python
W2 = Vt.T[:,:2] # Pegando PC1 e PC2
W2
```

```python
# Projetanndo X em 2D

X2D = X_centered.dot(W2)
X2D # Conjunto reduzido
```

```python
# Criando figura

fig = plt.figure(figsize=(10,7))
plt.scatter(X2D[:,0],X2D[:,1])
plt.title('Dataset Projetado no Plano')

plt.show()
```

# Usando Scikit-Learn para Calcular a PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X2D =  pca.fit_transform(X)
```

```python
# Criando figura

fig = plt.figure(figsize=(10,7))
plt.scatter(X2D[:,0],X2D[:,1])
plt.title('Dataset Projetado no Plano')

plt.show()
```
